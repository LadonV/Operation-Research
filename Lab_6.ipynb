```python

def get_ounces():
    if random.random() < 0.9:
        return 10
    else:
        return 9

def simulate_pulls_until_nine():
    pull_count = 0
    while True:
        pull_count += 1
        if get_ounces() == 9:
            return pull_count

pull_counts = []

for j in range(1000):
    pull_counts.append(simulate_pulls_until_nine())

average_pulls = sum(pull_counts) / 1000

print(f"{average_pulls:.2f}")
```

```python

plates = ['pink']*6 + ['nonpink']*18

pink_count_single = 0
pink_count_four = 0

for j in range(1000):
    random.shuffle(plates)

    draw1 = random.sample(plates, 1)
    if 'pink' in draw1:
        pink_count_single += 1

    draw4 = random.sample(plates, 4)
    if 'pink' in draw4:
        pink_count_four += 1

p_single = pink_count_single / 1000
p_four = pink_count_four / 1000

print(f"Estimated probability of drawing 1 pink plate: {p_single:.4f}")
print(f"Estimated probability of at least 1 pink in 4 draws: {p_four:.4f}")
```

```python
import random

def pull_until_pink(stack):
    drawn = []
    rejected = []

    while True:
        if not stack:
            break
        plate = stack.pop()
        drawn.append(plate)
        if plate == 'pink':
            break
        else:
            rejected.append(plate)

    stack.extend(rejected[::-1])
    return drawn, stack

plates_stack = ['pink']*6 + ['blue']*6 + ['black']*6 + ['green']*6
random.shuffle(plates_stack)

days = 1000
all_pink_count = 0

for day in range(days):
    drawn, plates_stack = pull_until_pink(plates_stack)

    top_six = plates_stack[-6:] if len(plates_stack) >= 6 else plates_stack
    if all(p == 'pink' for p in top_six):
        all_pink_count += 1

prob_all_pink_top = all_pink_count / days
print(f"Estimated fraction of days with all pink plates at top: {prob_all_pink_top:.4f}")
```
